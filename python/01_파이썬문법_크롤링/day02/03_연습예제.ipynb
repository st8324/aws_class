```python
# 문장을 입력받아 각 단어가 몇번 사용됐는지 판별하는 코드를 작성하세요
# 입력 : 안녕 내 이름 홍길동 네 이름 뭐니 내 이름 임꺽정 
# 출력
# 안녕 : 1 
# 내 : 2 
# 이름 : 3 
# 홍길동 : 1 
# 네 : 1 
# 뭐니 : 1 
# 임꺽정 : 1 

# 문장일 입력 
str = input("문장을 입력하세요 : ")


def analyze_words(sentence):
    # 문장을 리스트로 변환 
    words = sentence.split() 
    
    # 딕셔너리 변수 선언 
    dic = {} 
    
    # 반복문 : 리스트에 있는 값을 하나씩 꺼냄 
    for word in words:
        
        # 꺼낸 단어가 딕셔너리에 있는지 확인해서 있으면 1증가
        if word in dic:
            dic[word] += 1
        # 없으면 딕셔너리에 추가 후 값을 1로 설정 
        else:
            dic[word] = 1
    return dic

def analyze_words2(sentence):
    # 문장을 리스트로 변환 
    words = sentence.split() 
    # 딕셔너리 변수 선언 
    dic = {} 
    # 반복문 : 리스트에 있는 값을 하나씩 꺼냄 
    for word in words:
        # 꺼낸 단어가 딕셔너리에 없으면 추가(개수와 함께) 
        if word not in dic:
            dic[word] = words.count(word)
    return dic

# 딕셔너리를 출력하는 함수 
def print_dic(dic) : 
    for key, value in dic.items(): 
        print(f"\"{key}\" : {value}")

print_dic(analyze_words2(str))
```
