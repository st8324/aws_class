```python
# 홍길동 학생의 국어, 영어, 수학 성적을 콘솔 입력받아 numpy 배열로 만들고 콘솔에 출력하세요
# 콘솔에서 입력받을 때 사용할 문구 => 성적을 입력하세요 : 
import numpy as np

subjects = ["국어", "영어", "수학"]
list = []
# 과목수만큼 반복문
for subject in subjects:
	# 성적을 입력
	num = input(f"{subject} 성적을 입력하세요 : ")
	# 리스트에 추가
	list.append(int(num))
# 리스트를 numpy 배열로 변환
scores = np.array(list)
print(scores)
```

```python
# 학생의 평균을 출력하세요 
print(f"평균 : {scores.mean()}")
print(f"평균 : {np.mean(scores)}")
```
