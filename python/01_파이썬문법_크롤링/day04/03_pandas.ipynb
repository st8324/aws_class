```python
# 판다스 예제
import pandas as pd
```

```python
# 리스트를 이용 => Series
list = [100, 20, 30, 40]
scores = pd.Series(list, index=["국어", "영어", "수학", "과학"])

print(scores)
# 딕셔너리를 이용 => DataFrame
scores_dic = {
	'과목' : ["국어", "영어", "수학", "과학"] ,
	'성적' : [100, 90, 80, 70],
	'출석' : [100, 100, 0, 80],
	'수행' : [90, 100, 90, 30]
}
df = pd.DataFrame(scores_dic)
print(df)
```

```python
# 데이터 탐색 예제
print(f"상위 2행 : \n{df.head(2)}")
print("-"*20)
print(f"하위 2행 : \n{df.tail(2)}")
print("-"*20)
print(f"info : \n{df.info()}")
print("-"*20)
print(f"describe : \n{df.describe()}")
print("-"*20)
print(f"shape : \n{df.shape}")
print("-"*20)
```

```python
# 절취선
def bar():
	print("-"*30)
```

```python

# 데이터 선택 및 필터링
print("데이터 프레임")
bar()
print(df)
bar()

# 과목들 
print(df['과목'])
bar()

# 국어 정보 선택
print(f"국어 정보 : \n{df.loc[0]}")
bar()

# 과목의 수행평가 점수 조회
print("과목과 수행 평가 조회")
# 조회하려는 열이 여러개인 경우 열 목록을 리스트로 전달
print(df[['과목', '수행']])
bar()

# 국어, 수학 정보 조회
print("국어, 수학 정보 조회")
print(df.loc[[0,2]])
bar()

# 국어, 영어, 수학 정보 조회
print("국어, 영어, 수학 정보 조회")
print(df.iloc[0:3])
bar()
# 국어 성적
print(f"국어 성적 : {df.loc[0, '성적']}")
bar()
# 성적이 80이상인 데이터를 조회
print(df[df['성적']>= 80])
print(df.loc[df['성적']>= 80])
bar()
# 성적이 80이상인 과목들을 조회 
print(df[df['성적']>= 80]['과목'])
print(df.loc[df['성적']>= 80, '과목'])
print(df.query('성적 >= 80')['과목'])
```

```python
# 데이터 결합 및 그룹 예제
df3 = pd.DataFrame({
	"이름" : ["홍길동", "고길동", "둘리"], 
	"국어성적" : [100,90, 80]
})
df4 = pd.DataFrame({
	"이름" : ["홍길동", "고길동", "둘리"], 
	"영어성적" : [80,70, 90]
})
```

```python
# 국어 성적과 영어 성적 합치기 
total = pd.merge(df3, df4, on="이름")
print(total)
```

```python
# 동일한 형태(열 이름이 동일)의 두 테이블 합치기(위, 아래로)
total2 = pd.DataFrame({
	"이름" : ["개똥이", "하니", "또치"], 
	"국어성적" : [50,50, 90],
	"영어성적" : [30,40, 90],
})
total = pd.concat([total, total2])
print(total)
```

```python
# 반 열을 추가
total['반'] = 1
print(total)
```

```python
# 이름이 3자 미만인 학생은 2반으로 수정
total.loc[total['이름'].str.len() < 3, '반'] = 2
print(total)
```

```python
# 반 별 국어 성적 평균
print(total.groupby("반")['국어성적'].mean())
# 반 별 국어 성적 평균, 총합
print(total.groupby("반")['국어성적'].agg(['mean','sum']))
```

```python
# 반별로 정렬
total = total.sort_values(by='반')

# index 초기화 
total = total.reset_index(drop=True) # drop=True는 기존 인덱스를 삭제
print(total)
```

```python
# 중복 데이터를 마지막에 추가
# total.loc[6] = ["또치", 90, 90, 2]
total.loc[len(total)] = ["또치", 90, 90, 2]

# 중복 제거
print(total.drop_duplicates())
```
