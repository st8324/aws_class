```python
from selenium import webdriver
import time

# 브라우저가 바로 닫히는 것을 방지하는 옵션
options = webdriver.ChromeOptions()
options.add_experimental_option("detach", True)

# 브라우저 실행
driver = webdriver.Chrome(options=options)

url = "https://www.naver.com"
# 페이지 이동
driver.get(url)

# 페이지 로딩을 위해 5초 대기 : 충분한 로딩시간
time.sleep(5)

# html 코드를 읽어옴
html_source = driver.page_source

# 요소 탐색 및 동작 가능
# 예시 생략
    

with open('naver.html','wt', encoding='utf-8') as f:
	f.write(html_source)

# 모든 작업 완료 후 브라우저 종료
driver.quit()
```
