```python
# лҚ°мқҙн„° мғқм„ұ : csvнҢҢмқјмқ„ л¶Ҳлҹ¬мҳҙ
import pandas as pd

df = pd.read_csv('../../data/csv/sentiment_data_long.csv')
```

```python
# лҸ…лҰҪ, мў…мҶҚліҖмҲҳ л¶„лҰ¬
X = df['sentence']
y = df['label']
```

```python
# нӣҲл Ё, н…ҢмҠӨнҠё л¶„лҰ¬ 
DATA_SIZE = 1000
TRAIN_SIZE = int(DATA_SIZE * 0.8)

X_train, X_test = X[:TRAIN_SIZE], X[TRAIN_SIZE: ]
y_train, y_test = y[:TRAIN_SIZE], y[TRAIN_SIZE: ]
```

```python
# нҳ•нғңмҶҢ л¶„лҰ¬ мӢң лӘЁл“  нҳ•нғңмҶҢлҘј нҸ¬н•Ё
# л¶„лҰ¬н•ң лӢЁм–ҙл“Өмқ„ н•©м№Ё
from konlpy.tag import Okt

def get_preprocessing(sentence):
	okt = Okt()
	result = okt.pos(sentence, stem=True) # л¬ёмһҘмқ„ нҳ•нғңмҶҢлі„лЎң лӮҳлҲ”. лӢЁ, мӣҗнҳ•мңјлЎң
	words = [word for word, pos in result]
	return " ".join(words)

# X_trainкіј X_testмқҳ мһҲлҠ” л¬ёмһҘл“Өмқ„ get_preprocessingмқ„ м Ғмҡ©
X_train = X_train.apply(get_preprocessing)
X_test = X_test.apply(get_preprocessing)
```

```python
# лІЎн„°нҷ” 
from tensorflow.keras import layers, models
vectorize_layer = layers.TextVectorization(
	max_tokens=1000,
	output_mode="int",
	output_sequence_length=10
)
vectorize_layer.adapt(X_train.tolist())
```

```python
import tensorflow as tf

# лӘЁлҚё м„Өкі„
model = models.Sequential([
	layers.Input((1, ), dtype=tf.string),
	vectorize_layer, 
	layers.Embedding(input_dim=1000, output_dim=64),
	layers.LSTM(32),
	layers.Dense(1, activation='sigmoid')#м¶ңл Ҙмёө
])
```

```python
model.compile(
	optimizer='adam', 
	loss='binary_crossentropy', 
	metrics=['accuracy']
) 
```

```python
history = model.fit(X_train.values, y_train, epochs=50, verbose=1)
```

```python
# нҸүк°Җ
_, acc = model.evaluate(X_test.values, y_test)
print(f'м •нҷ•лҸ„ : {acc}')
```
