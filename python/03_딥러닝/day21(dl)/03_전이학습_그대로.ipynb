```python
# н…ҢмҠӨнҠён•  мқҙлҜём§Җ мӨҖл№„
import requests
from tensorflow.keras.preprocessing import image
from io import BytesIO

url = 'https://i.namu.wiki/i/fBf3c5Ty1j4AaTkakEqu5xAEF1vAunGZ2ClshnWW1tpcYZWbSMxT0UQKPGQxYJKX03y5wJSnDclukDTr2XJbWTqurju6dTKDm1CzZNo883jax_WadPTMmfTfRoEjugdKDBLGEJVZpdRe7q51M0-NxQ.webp'

response = requests.get(url)
img = image.load_img(BytesIO(response.content), target_size=(224,224))
```

```python
import numpy as np
# мқҙлҜём§Җ м „мІҳлҰ¬ => мһ…л Ҙ нҳ•нғң л§һм¶”кё°, мқҙлҜём§ҖлҘј л°°м—ҙлЎң
X = image.img_to_array(img)
print(X.shape)
X = np.expand_dims(X, axis=0)
print(X.shape)
```

```python
# MobileNetV2м „мҡ© м „мІҳлҰ¬ 
X = mobilenet_v2.preprocess_input(X)
# print(X)
```

```python
# мҳҲмёЎ
p = model.predict(X)
```

```python
# кІ°кіјлҘј н•ҙм„қ
results = mobilenet_v2.decode_predictions(p, top=3)[0]

print('-'*30)
for i, (_, label, prob) in enumerate(results):
	print(f'{i+1}. {label} : {prob*100:.2f}%')
print('-'*30)
```

```python
import numpy as np
import requests
from tensorflow.keras.applications import mobilenet_v2
from tensorflow.keras.preprocessing import image
from io import BytesIO

def predict_img_url(url):
	# лӘЁлҚё л¶Ҳлҹ¬мҳӨкё°. л¶„лҘҳкё° нҸ¬н•Ё(1000к°ңмқҳ л¶„лҘҳ)
	model = mobilenet_v2.MobileNetV2(weights='imagenet', include_top=True)
	
	# мқҙлҜём§Җ м •ліҙлҘј к°Җм ёмҳҙ
	response = requests.get(url)
	# к°Җм ёмҳЁ мқҙлҜём§ҖлҘј л°”мқҙнҠё нҳ•нғңлЎң ліҖнҷҳ
	img = image.load_img(BytesIO(response.content), target_size=(224,224))

	# мқҙлҜём§Җ м „мІҳлҰ¬ => мһ…л Ҙ нҳ•нғң л§һм¶”кё°, мқҙлҜём§ҖлҘј л°°м—ҙлЎң
	X = image.img_to_array(img)
	X = np.expand_dims(X, axis=0)

	# MobileNetV2м „мҡ© м „мІҳлҰ¬ 
	X = mobilenet_v2.preprocess_input(X)
	
	# мҳҲмёЎ
	predicts = model.predict(X)

	# кІ°кіјлҘј н•ҙм„қ
	results = mobilenet_v2.decode_predictions(predicts, top=3)[0]

	print('-'*30)
	for i, (_, label, prob) in enumerate(results):
		print(f'{i+1}. {label} : {prob*100:.2f}%')
	print('-'*30)
	
```

```python
# л¶ҲлҸ… мқҙлҜём§ҖлЎң н…ҢмҠӨнҠё
# 1. л¶ҲлҸ…
# 2. л¶Ҳл§ҲмҠӨнӢ°н”„
# 3. мҠӨнғҖнҸ¬л“ңм…” л¶Ҳ н…ҢлҰ¬м–ҙ
predict_img_url('https://i.namu.wiki/i/fBf3c5Ty1j4AaTkakEqu5xAEF1vAunGZ2ClshnWW1tpcYZWbSMxT0UQKPGQxYJKX03y5wJSnDclukDTr2XJbWTqurju6dTKDm1CzZNo883jax_WadPTMmfTfRoEjugdKDBLGEJVZpdRe7q51M0-NxQ.webp')
```
