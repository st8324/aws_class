# 함수 정리

```python
# 매개변수의 기본값 
# min부터 max까지 정수를 출력하는 함수 
# min부터 max까지 출력하는데, max가 없으면 10까지 출력되도록 함 
# 단, 기본값은 뒤에서부터 채워야 함 
def printNums(min, max = 10):
    for i in range(min, max+1):
        print(i)
# min이라는 변수에 1을, max라는 변수에 3을 저장해서 실행하는 것이 아니라 
# 매개변수 min에 1을, 매개변수 max에 3을 저장해서 실행함
# 키워드 인자를 사용하지 않으면 순서가 바뀌면 실행 결과가 달라질 수 있음
print("====================")
print("print(1,3)")
printNums(1,3) 
print("print(3,1)")
printNums(3,1) 
print("====================")
# 키워드 인자를 사용하면 순서가 바뀌어도 실행 결과가 동일 
print("printNums(min = 1, max = 3)")
printNums(min = 1, max = 3)
print("printNums(max = 3, min = 1)")
printNums(max = 3, min = 1)
print("====================")
# printNums(1)
```

```python
# 함수의 가변 인자 => 매개변수가 개수가 고정이 아님 
# 정수의 합을 구하는 함수 
def sumNums(*args):
    sum = 0
    for num in args:
        sum += num 
    return sum 

print(sumNums(1,3,5,7,9))
print(sumNums(2,4,6,8,10))

# 정보를 이용하여 객체를 반환하는 함수 
def createDic(**kwargs):
    dic = {}
    # { "name" : "홍길동", "age" : 21, "role" : "user"}
    # { "title" : "쉬운 파이썬", "price" : 30000, "author" : "고길동"}
    for key, value in kwargs.items(): 
        dic[key] = value
    return dic 
# 키워드 인자를 이용하여 호출 
user = createDic(name="홍길동", age=21, role="user")
print(user)
book = createDic(title="쉬운 파이썬", price=30000, author ="고길동")
print(book)
```

```python
# 람다 함수  
# 두 정수의 합을 알려주는 함수 
add = lambda num1, num2 : num1 + num2 

print(add(1,2))

# 홀짝 판별하는 함수 
printEven = lambda num : print(f"{num}은 {"짝수" if num % 2 == 0 else "홀수"}")
printEven(10) 
printEven(11)
```
