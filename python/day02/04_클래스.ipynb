```python
# 2차원 좌표평면의 점을 나타내는 Point 클래스를 선언 

# Point 클래스의 객체를 생성
class Point: 
    # 클래스 변수
    name : str = "포인트 클래스" 

    # 생성자 
    def __init__(self, x:int = 0, y:int = 0): 
        # 인스턴스 변수 
        self.__x = x 
        self.__y = y

    # 함수
    def print(self):
        print(f"({self.__x}, {self.__y})")
    # __x의 getter
    @property 
    def x(self): 
        return self.__x
    # __x의 setter
    @x.setter
    def x(self, x): 
        self.__x = x 

    @property
    def y(self): 
        return self.__y 
    @y.setter
    def y(self, y):
        self.__y = y 

print(Point.name)
p1 = Point(0,0)
p2 = Point(1,1) 
p3 = Point() 
p1.print()
p2.print()
p3.x = 100
p3.y = 100
p3.print()
```

```python
# 클래스 변수 확인 예제 
class A: 
    #클래스 변수 => 모든 객체가 같은 값을 가짐 
    name : str = "A클래스입니다" 
    def __init__(self, age):
        # 인스턴스 변수 => 객체마다 다른 값을 가짐 
        self.age = age

    def print(self):
        print(f"name : {A.name}")
        print(f"나이 : {self.age}")

a1 = A(10) 
a1.print()
a2 = A(20) 
a2.print()
A.name = "A클래스2입니다."
a1.age = 30
a1.print()
a2.print()
```

```python
# 은행 계좌 클래스 
# 소유자(owner), 잔액(balance) => 인스턴스변수 
# 기능 : 출금(withdraw), 입금(deposit) => 함수 
class BankAccount: 

    def __init__(self, 소유자 : str, 잔액 : int = 0): 
        self.__소유자 = 소유자 
        self.__잔액 = 잔액 

    def 출금(self, 출금액:int): 
        print(f"{출금액:,}원 출금 중 ...")
        if 출금액 <= 0: 
            print("출금액은 0원보다 커야 합니다.")
        elif(self.__잔액 < 출금액):
            print(f"잔액이 부족해서 출금할 수 없습니다. 현재 잔액 : {self.__잔액}")
        else:
            self.__잔액 -= 출금액 
            print(f"출금 후 계좌 잔액 : {self.__잔액:,}원")

    def 입금(self, 입금액:int): 
        print(f"{입금액:,}원 입금 중 ...")
        if 입금액 <= 0: 
            print("입금액은 0원보다 커야 합니다.")
        else : 
            # slef.__잔액 = self.__잔액 + 입금액
            self.__잔액 += 입금액
            print(f"입금 후 계좌 잔액 : {self.__잔액:,}원")

    def 잔액조회(self): 
        print(f"{self.__소유자}님 계좌 잔액 : {self.__잔액:,}원")

# 홍길동이 계좌 개설 
account = BankAccount("홍길동")
# 100000(십만) 입금 
account.입금(100000)
# 50000(5만) 출금 
account.출금(50000)
# 70000(7만) 출금 => 잔액 부족 알림 
account.출금(70000)
account.잔액조회()
```
