```python
# 예외 처리 예제 

num1, num2 = 1 , 0
try:
	print(f"{num1} / {num2} = {num1 / num2}")
except ZeroDivisionError:
	print("0으로 나눌 수 없습니다")
else:
	print("나누기 성공!!")
finally:
	print("예제를 종료합니다.")

print("----------------------")

# 예외 처리 없이 예외를 조건문으로 다룸
if(num2 == 0):
	print("0으로 나눌 수 없습니다")
else:
	print(f"{num1} / {num2} = {num1 / num2}")
	print("나누기 성공!!")
print("예제를 종료합니다.")
```

```python
# 다양한 예외를 처리 
try:
	num1 = int(input("정수1 입력 : "))
	num2 = int(input("정수2 입력 : "))
	print(f"{num1} / {num2} = {num1 / num2}")
except ValueError:
	print("숫자만 입력하세요")
except ZeroDivisionError:
	print("0으로 나눌 수 없습니다.")
except Exception as e:
	print(f"예상치 못한 예외 발생!! : {e}")
```

```python
# 예외 발생 시키기 예제
class User: 

	def __init__(self, name, age):
		self.__name = name 
		if age < 0:
			raise ValueError("나이는 0이상이어야 합니다.")
		self.__age = age 
	
	def info(self):
		print(f"이름 : {self.__name}, 나이 : {self.__age}")
	# age의 getter
	@property
	def age(self):
		return self.__age
	# age의 setter
	@age.setter
	def age(self, age):
		if age < 0: 
			raise ValueError("나이는 0이상이어야 합니다.")
		self.__age = age

try:
	users = [
		User("홍길동", 21),
		User("둘리", 0)
	]

	users[1].age = -2

	for user in users:
		user.info()
except Exception as e:
	print(f"예외 발생 : {e}")
```
