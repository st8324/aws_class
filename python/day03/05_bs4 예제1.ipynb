```python
# 네이버 메인페이지 가져오기

import requests
from bs4 import BeautifulSoup

# sample_site.html 파일을 우클릭 > Open with Live Server로 실행 후 실행
url = "http://127.0.0.1:5500/day03/sample_site.html"

headers = {
	'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/146.0.0.0 Safari/537.36'
}

response = requests.get(url, headers=headers, timeout=5)

try:
	response.raise_for_status()
	print("연결 성공!")

	soup = BeautifulSoup(response.text, "lxml")
	# title 태그
	title_tag = soup.title
	#body 태그
	body_tag = soup.body
	
	#body 태그 안에 있는 모든 li태그들 => 태그로 검색
	list = body_tag.find_all("li")

	# li태그 내용을 출력
	for li in list : 
		print(li.text)

	#body 태그 안에 있는 모든 li태그들 => 선택자로 검색
	list = soup.select("ul>li")

	for li in list : 
		print(li.text)
		# 클래스 정보 가져오기
		print(f"클래스명 : {li.get("class")}")

except Exception as e:
	print(f"예외 발생 : {e}")
```
