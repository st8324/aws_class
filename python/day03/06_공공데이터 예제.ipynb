```python
# 서울특별시 법정동 정보를 모두 가져오는 예제
import requests
from bs4 import BeautifulSoup
import time

def get_stan_regin_list(page : int = 1, num_or_rows : int = 10, locatadd_nm : str = ""):
	"""법정동 정보를 가져와서 리스트로 반환"""

	url = "https://apis.data.go.kr/1741000/StanReginCd/getStanReginCdList"
	params = {
		'ServiceKey' : '내서비스키',
		'pageNo' : page,
		'numOfRows' : num_or_rows,
		'type' : 'xml'
	}
	if str : 
		params['locatadd_nm'] = locatadd_nm

	try:
		response = requests.get(url, params=params)
		
		response.raise_for_status()
		
		soup = BeautifulSoup(response.text, "lxml-xml")
		
		# 총 개수 
		# total_count = soup.find("totalCount").text
		# print(f"서울특별시 전체 법정동 수 : {total_count}")
		
		# 실제 가져온 데이터들 
		rows = soup.find_all("row")

		data = []

		for row in rows:
			region_cd = row.find("region_cd").text # 지역코드
			sido_cd = row.find("sido_cd").text # 시도코드 
			locatadd_nm = row.find("locatadd_nm").text # 지역주소명
			locallow_nm = row.find("locallow_nm").text # 최하위 지역명 
			data.append({
				'region_cd' : region_cd, 
				'sido_cd' : sido_cd, 
				'locatadd_nm' : locatadd_nm, 
				'locallow_nm' : locallow_nm
			})
		
		return data
		
	except Exception as e:
		print(f"예외 발생 : {e}")
		return []

def get_stan_regin_total_count(locatadd_nm : str = ""):
	"""법정동 전체 개수를 조회"""

	url = "https://apis.data.go.kr/1741000/StanReginCd/getStanReginCdList"
	params = {
		'ServiceKey' : '서비스키',
		'pageNo' : 1,
		'numOfRows' : 10,
		'type' : 'xml'
	}
	if str : 
		params['locatadd_nm'] = locatadd_nm

	try:
		response = requests.get(url, params=params)
		
		response.raise_for_status()
		
		soup = BeautifulSoup(response.text, "lxml-xml")
		
		# 총 개수 
		total_count = soup.find("totalCount").text
		# print(f"서울특별시 전체 법정동 수 : {total_count}")
		
		return int(total_count)
		
	except Exception as e:
		print(f"예외 발생 : {e}")
		return 0
# 시작 페이지
page = 1
# 전체 개수
total_count = get_stan_regin_total_count()
# 한페이지에 조회할 개수 
num_or_rows = 10
# 최대 페이지 
max_page = total_count // num_or_rows
# 예로 431개이면 43페이지가 아니라 44페이지어야 함
max_page = max_page if max_page * num_or_rows >= total_count else max_page + 1

# 최종으로 넣을 데이터
data = []

for i in range(1, 10 + 1): # 원래는 max인데 10으로 임시 수정 => max로하면 시간 오래 걸림
	tmp = get_stan_regin_list(page=page, num_or_rows=num_or_rows)
	data.extend(tmp)
	print(f"{i}페이지 로딩 완료! ")
	time.sleep(1)

print(len(data))
print(data)
```
