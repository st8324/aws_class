```python
# 판다스 예제
import pandas as pd
```

```python
# 리스트를 이용 => Series
list = [100, 20, 30, 40]
scores = pd.Series(list, index=["국어", "영어", "수학", "과학"])

print(scores)
# 딕셔너리를 이용 => DataFrame
scores_dic = {
	'과목' : ["국어", "영어", "수학", "과학"] ,
	'성적' : [100, 90, 80, 70],
	'출석' : [100, 100, 0, 80],
	'수행' : [90, 100, 90, 30]
}
df = pd.DataFrame(scores_dic)
print(df)
```

```python
# 데이터 탐색 예제
print(f"상위 2행 : \n{df.head(2)}")
print("-"*20)
print(f"하위 2행 : \n{df.tail(2)}")
print("-"*20)
print(f"info : \n{df.info()}")
print("-"*20)
print(f"describe : \n{df.describe()}")
print("-"*20)
print(f"shape : \n{df.shape}")
print("-"*20)
```

```python
# 절취선
def bar():
	print("-"*30)
```

```python

# 데이터 선택 및 필터링
print("데이터 프레임")
bar()
print(df)
bar()

# 과목들 
print(df['과목'])
bar()

# 국어 정보 선택
print(f"국어 정보 : \n{df.loc[0]}")
bar()

# 과목의 수행평가 점수 조회
print("과목과 수행 평가 조회")
# 조회하려는 열이 여러개인 경우 열 목록을 리스트로 전달
print(df[['과목', '수행']])
bar()

# 국어, 수학 정보 조회
print("국어, 수학 정보 조회")
print(df.loc[[0,2]])
bar()

# 국어, 영어, 수학 정보 조회
print("국어, 영어, 수학 정보 조회")
print(df.iloc[0:3])
bar()
# 국어 성적
print(f"국어 성적 : {df.loc[0, '성적']}")

```
