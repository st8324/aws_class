```python
# 샘플 데이터를 학습해서 30도일 때 아이스크림 판매량을 예측하세요.
# 그래프도 출력
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
# 데이터 설정
dataset = pd.read_csv('선형회귀_아이스크림.csv')
```

```python
# 독립변수, 종속변수 분리
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
```

```python
# 훈련데이터, 테스트데이터 분리
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=24)
```

```python
# 학습
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# 예측
y_pred = model.predict(X_test)
```

```python
# 온도가 다음과 같을 때 판매량 예측
tem_list = [[11], [13], [14]]
for tem in tem_list:
	y_res = model.predict([tem])
	print(f'{tem}도일 때 판매량(예측) : {y_res.round(0)}개')
```

```python
plt.scatter(X_train, y_train, color='blue') # 훈련 데이터를 점으로 출력
plt.plot(X_train, model.predict(X_train), color='red') # 학습을 통해 나온 직선
plt.scatter(X_test, y_test, color='green') # 테스트 데이터의 실제 데이터
plt.title('score by hour') # 그래프 제목
plt.xlabel('hour')# x축 => 독립변수
plt.ylabel('score')# y축 => 종속변수
plt.show() # 그래프 출력
```
