```python
# 그리드 서치 예제

from sklearn.datasets import load_breast_cancer

# 데이터로드
dataset = load_breast_cancer()
X, y = dataset['data'], dataset['target']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

# 테스트할 파라미터 딕셔너리
param_grid = {
    'C': [1, 10], # 오차를 얼마나 허용할지를 결정
    'gamma': [0.001], # 한 데이터가 결정 경계형성에 미치는 영향 범위
    'kernel': [ 'linear'] # 복잡한 곡선이나 원형
}

# GridSearchCV 객체 생성 (모델, 파라미터, 교차검증 횟수 설정)
# n_jobs=-1은 모든 CPU 코어를 사용하여 병렬 처리를 하겠다는 뜻입니다.
grid_search = GridSearchCV(SVC(), param_grid, cv=5, n_jobs=-1, verbose=2)

# 모든 조합에 대해 학습 시작
grid_search.fit(X_train, y_train)

# 4. 결과 확인
print(f"최적의 파라미터: {grid_search.best_params_}")
print(f"최고 검증 점수: {grid_search.best_score_:.4f}")

# 최적의 모델로 바로 예측하기
best_model = grid_search.best_estimator_
prediction = best_model.predict(X_test)
```
