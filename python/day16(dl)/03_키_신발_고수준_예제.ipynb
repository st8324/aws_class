```python
# 키 데이터와 실발 데이터를 이용하여 학습하고, 키 데이터를 이용하여 신발 사이즈를 예측하는 예제
# 고수준으로 작성 => Keras 모델 이용
import tensorflow as tf
키데이터 = tf.constant([170, 180, 175, 160], dtype=tf.float32)
신발데이터 = tf.constant([260, 270, 265, 255], dtype=tf.float32)
```

```python
# 모델 설계 
model = tf.keras.Sequential([
	# 레이어를 작성
	# 마지막 레이어는 출력층
	# 1은 해당 레이어를 지나면 결과가 몇개 나와야 하는지를 결정
	# 지금은 신발 사이즈 값 1개만 나오면 되니까 1
	# input_shape(1,) : 입력층의 형태. 입력층은 키 1개이니까 1,
	tf.keras.layers.Dense(1, input_shape=(1,))
])

# 모델 설정. 옵티마이저를 어떤 모델로 할건지, 오차는 어떤 방식으로 계산할건지를 설정
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.1), loss='mse')
```

```python
# 학습 실행
model.fit(키데이터, 신발데이터, epochs=1000, verbose=0)
```

```python
import numpy as np
측정키 = 173
# 텐서 모델들은 예측할 때, numpy 리스트가 필요
예측신발 = model.predict(np.array([[측정키]], dtype='float32'))
print(f'측정키 : {측정키}, 예측신발사이즈 : {예측신발[0][0]:.2f}')
```
