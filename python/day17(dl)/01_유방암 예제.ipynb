```python
# 데이터 생성. 유방암 sklean
from sklearn.datasets import load_breast_cancer

dataset = load_breast_cancer()
```

```python
# 독립, 종속 분리
X = dataset.data
y = dataset.target
```

```python
# 훈련, 테스트 분리
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# 정규화
# import tensorflow as tf
from tensorflow.keras import layers, models
nomalizer = layers.Normalization(axis=-1)
nomalizer.adapt(X_train)
```

```python
# 모델 설계. 입력층, 은닉층, 출력층
# 각 층마다 활성화 함수(입력층 제외)
model = models.Sequential([
	nomalizer, 
	layers.Dense(64, activation='relu'),
	layers.Dense(32, activation='relu'),
	layers.Dense(16, activation='relu'),
	layers.Dense(1, activation='sigmoid') # 출력층
])
```

```python
# 모델 설정 : 손실함수, 옵티마이저, 평가지표
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# 학습
history = model.fit(X_train, y_train, epochs=1000, batch_size=50, verbose=0)
```

```python
# 평가
_ , test_acc = model.evaluate(X_test, y_test)
print(f'정확도 : {test_acc}')
```

```python
# 예측 : 실제 데이터 중 하나의 값을 가져와서 진행
import numpy as np
idx = 1
sample = [ 
	x + 0.001 if x < 5 else (
	x + 0.5 if x < 20 else (
	x + 3 if x < 100 else (
	x + 10)))
	for x in X_test[idx] # 0번지가 암(1)
]
print(sample)
prediction = model.predict(np.array([sample]))
print(f'예상 결과 : {y_test[idx]}')
print(f'예측 결과 : {prediction.round(4)}') # 0.000014908
```
