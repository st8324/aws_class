```python
import pandas as pd
import random

def generate_contrastive_dataset(file_name="sentiment_train_data.csv", count=1000):
    # кёҚм •м Ғмқё мҡ”мҶҢл“Ө
    pos_elements = [
        "мҳҒмғҒлҜёк°Җ м••лҸ„м ҒмңјлЎң нҷ”л Өн•ҳкі ", "л°°мҡ°л“Өмқҳ лӘ°мһ…к°җмқҙ м—„мІӯлӮҳл©°", 
        "мқҢм•… м„ кіЎмқҙ к°ҖмҠҙмқ„ мҡёлҰ¬кі ", "м—°м¶ңл Ҙмқҙ м„ёкі„м Ғмқё мҲҳмӨҖмқҙкі ", 
        "мҙҲл°ҳл¶Җ м „к°ңк°Җ л§Өмҡ° нқҘлҜём§„м§„н•ҙм„ң"
    ]
    # л¶Җм •м Ғмқё мҡ”мҶҢл“Ө
    neg_elements = [
        "мҠӨнҶ лҰ¬к°Җ л„Ҳл¬ҙ к°ңм—°м„ұмқҙ м—Ҷкі ", "кІ°л§җмқҙ н—Ҳл¬ҙн•ҳкё° м§қмқҙ м—Ҷмңјл©°", 
        "м§ҖлЈЁн•ң мһҘл©ҙмқҙ л°ҳліөлҗҳм–ҙм„ң", "кё°лҢҖн–ҲлҚҳ кІғліҙлӢӨ мӢӨл§қк°җмқҙ нҒ¬кі ", 
        "лҸҲкіј мӢңк°„мқҙ м•„к№қлӢӨлҠ” мғқк°Ғмқҙ л“Өм–ҙм„ң"
    ]
    
    # л°ҳм „ м ‘мҶҚмӮ¬
    to_neg_conj = [" н•ҳм§Җл§Ң кІ°кіјм ҒмңјлЎң ", " к·ёлҹ¬лӮҳ кІ°көӯм—җлҠ” ", " к·ёлҹјм—җлҸ„ л¶Ҳкө¬н•ҳкі  "]
    to_pos_conj = [" н•ҳм§Җл§Ң мқҳмҷёлЎң ", " к·ёлҹ¬лӮҳ мӢ кё°н•ҳкІҢлҸ„ ", " к·ёлҹјм—җлҸ„ л¶Ҳкө¬н•ҳкі  "]

    # мөңмў… к°җм • н‘ңнҳ„
    final_neg = ["мөңм•…мқҳ мҳҒнҷ”лқјлҠ” нҸүк°ҖлҘј н”јн•  мҲҳ м—ҶлӢӨ.", "м ҲлҢҖ м¶”мІңн•ҳкі  мӢ¶м§Җ м•ҠмқҖ мһ‘н’ҲмқҙлӢӨ.", "мӢӨл§қмҠӨлҹ¬мҡҙ мҷ„м„ұлҸ„м—җ нҷ”к°Җ лӮ  м •лҸ„лӢӨ."]
    final_pos = ["мқёмғқ мөңкі мқҳ лӘ…мһ‘мңјлЎң лӮЁмқ„ кІғ к°ҷлӢӨ.", "л°ҳл“ңмӢң к·№мһҘм—җм„ң лҙҗм•ј н•  мҲҳмһ‘мқҙлӢӨ.", "к№ҠмқҖ м—¬мҡҙмқ„ лӮЁкё°лҠ” мҷ„лІҪн•ң мҳҒнҷ”лӢӨ."]

    dataset = []

    # л¶Җм • л¬ёмһҘ 500к°ң (м•һмқҖ м№ӯм°¬, л’ӨлҠ” л№„нҢҗ)
    for _ in range(count // 2):
        front = " ".join(random.sample(pos_elements, 3))
        sentence = front + random.choice(to_neg_conj) + random.choice(final_neg)
        dataset.append([sentence, 0]) # 0: л¶Җм •

    # кёҚм • л¬ёмһҘ 500к°ң (м•һмқҖ л№„нҢҗ, л’ӨлҠ” м№ӯм°¬)
    for _ in range(count // 2):
        front = " ".join(random.sample(neg_elements, 3))
        sentence = front + random.choice(to_pos_conj) + random.choice(final_pos)
        dataset.append([sentence, 1]) # 1: кёҚм •

    # лҚ°мқҙн„°н”„л Ҳмһ„ мғқм„ұ л°Ҹ CSV м ҖмһҘ
    df = pd.DataFrame(dataset, columns=['text', 'label'])
    df.to_csv(file_name, index=False, encoding='utf-8-sig')
    print(f"{file_name} нҢҢмқјмқҙ мғқм„ұлҗҳм—ҲмҠөлӢҲлӢӨ.")

# лҚ°мқҙн„° мғқм„ұ мӢӨн–ү
generate_contrastive_dataset()
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# лҚ°мқҙн„° лЎңл“ң л°Ҹ м „мІҳлҰ¬
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train = x_train.reshape(-1, 784) / 255.0
x_test  = x_test.reshape(-1, 784) / 255.0

# лӘЁлҚё м •мқҳ
model = tf.keras.Sequential([
    tf.keras.layers.Dense(256, activation='relu', input_shape=(784,)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64,  activation='relu'),
    tf.keras.layers.Dense(10,  activation='softmax')
])

# м»ҙнҢҢмқј
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

# мҪңл°ұ
callbacks = [
    tf.keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True),
    tf.keras.callbacks.ModelCheckpoint('mnist_best.keras', save_best_only=True)
]

# н•ҷмҠө
history = model.fit(
    x_train, y_train,
    epochs=50,
    batch_size=128,
    validation_split=0.1,
    callbacks=callbacks,
    verbose=1
)

# нҸүк°Җ
loss, acc = model.evaluate(x_test, y_test, verbose=0)
print(f"\nрҹҺҜ н…ҢмҠӨнҠё м •нҷ•лҸ„: {acc:.4f}")

# кІ°кіј мӢңк°Ғнҷ”
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Val')
plt.title('м •нҷ•лҸ„'); plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Val')
plt.title('мҶҗмӢӨ'); plt.legend()
plt.show()
```

```python
from tensorflow.keras import layers, models

model = models.Sequential([
  # 1. мһ…л Ҙ к·ңкІ© м •мқҳ
  layers.Input(shape=(32, 32, 3)),
  
  # 2. лҚ°мқҙн„° мҰқк°• (н•ҷмҠө мӢңм—җл§Ң мһ‘лҸҷ)
  layers.RandomFlip("horizontal"),
  layers.RandomRotation(0.1),
  
  # 3. м •к·ңнҷ” (н•ҷмҠө/мҳҲмёЎ кіөнҶө мһ‘лҸҷ)
  layers.Rescaling(1./255),
  
  # 4. CNN нҠ№м§• м¶”м¶ңл¶Җ
  layers.Conv2D(32, (3, 3), activation='relu'),
  layers.MaxPooling2D((2, 2)),
  
  # 5. л¶„лҘҳкё° (Dense)
  layers.Flatten(),
  layers.Dense(64, activation='relu'),
  layers.Dense(10, activation='softmax')
])
```

```python
import tensorflow as tf

texts = [
    "лӮҳлҠ” мқҙ мҳҒнҷ”к°Җ м •л§җ мўӢм•ҳлӢӨ",
    "мөңм•…мқҳ мҳҒнҷ”мҳҖлӢӨ мӢңк°„ лӮӯл№„",
    "л°°мҡ°л“Өмқҳ м—°кё°к°Җ нӣҢлҘӯн–ҲлӢӨ"
]

# Tokenizer м„Өм •
tokenizer = tf.keras.preprocessing.text.Tokenizer(
    num_words=10000,    # мғҒмң„ 10,000 лӢЁм–ҙл§Ң мӮ¬мҡ©
    oov_token="<OOV>"  # лҜёл“ұлЎқ лӢЁм–ҙ мІҳлҰ¬
)

tokenizer.fit_on_texts(texts)

# лӢЁм–ҙ вҶ’ м •мҲҳ ліҖнҷҳ
sequences = tokenizer.texts_to_sequences(texts)
print("мқёмҪ”л”©лҗң мӢңнҖҖмҠӨ:")
for t, s in zip(texts, sequences):
    print(f"  '{t}' вҶ’ {s}")

print("\\nлӢЁм–ҙ мӮ¬м „ (мқјл¶Җ):", dict(list(tokenizer.word_index.items())[:10]))
```
